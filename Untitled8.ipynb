### 全てのCSVファイルをPandas DataFrameとして読み込む

`os`モジュールを使用して`/content/`ディレクトリ内のCSVファイルを特定し、それぞれをPandas DataFrameとして読み込みます。読み込んだDataFrameは辞書`dfs`にファイル名をキーとして格納します。その後、各DataFrameの先頭5行を表示します。

````python
import pandas as pd
import os

# CSVファイルが格納されているディレクトリ
directory = '/content/'

# DataFrameを格納する辞書
dfs = {}

# ディレクトリ内の全てのファイルをリストアップ
for filename in os.listdir(directory):
    if filename.endswith('.csv'):
        filepath = os.path.join(directory, filename)
        try:
            # CSVファイルを読み込み、DataFrameとして格納
            df_name = filename.replace('.csv', '') # .csv拡張子を除去してキーとして使用
            dfs[df_name] = pd.read_csv(filepath)
            print(f"DataFrame '{df_name}' を読み込みました。先頭5行を表示します:")
            display(dfs[df_name].head())
            print("\n" + "-"*50 + "\n") # 区切り線
        except Exception as e:
            print(f"ファイル {filename} の読み込み中にエラーが発生しました: {e}")

print("全てのCSVファイルの読み込みが完了しました。")
````

### 読み込んだデータの欠損値を確認

`dfs`辞書に格納されている各DataFrameについて、列ごとの欠損値の合計を表示します。これにより、どのDataFrameのどの列にデータがないかを把握できます。

````python
print("各DataFrameの欠損値の合計：")
for df_name, df in dfs.items():
    print(f"\n--- DataFrame '{df_name}' ---")
    missing_values = df.isnull().sum()
    if missing_values.sum() == 0:
        print("このDataFrameには欠損値がありません。")
    else:
        display(missing_values[missing_values > 0].sort_values(ascending=False).to_frame(name='Missing Values'))
````

### 欠損値の処理（例：`Sale Report`）

`Sale Report` DataFrameには、数値型とカテゴリ型の両方に欠損値があります。ここでは、それぞれの型に対して異なる戦略を適用する例を示します。

-   **数値型（`Stock`）**: 中央値で欠損値を埋めます。
-   **カテゴリ型（`SKU Code`, `Category`, `Color`, `Design No.`, `Size`）**: 最頻値または 'Unknown' で欠損値を埋めます。

````python
# Sale Report DataFrameの欠損値処理
sale_report_df = dfs['Sale Report'].copy()

print("--- Sale Report DataFrame 欠損値処理前 ---")
display(sale_report_df.isnull().sum()[sale_report_df.isnull().sum() > 0].to_frame(name='Missing Values Before'))

# 数値型の欠損値を中央値で埋める (例: 'Stock')
# 'Stock' が数値型であることを確認してから適用
if 'Stock' in sale_report_df.columns and pd.api.types.is_numeric_dtype(sale_report_df['Stock']):
    median_stock = sale_report_df['Stock'].median()
    sale_report_df['Stock'].fillna(median_stock, inplace=True)
    print(f"'Stock' 列の欠損値を中央値 {median_stock} で埋めました。")

# カテゴリ型の欠損値を最頻値または 'Unknown' で埋める
for col in ['SKU Code', 'Category', 'Color', 'Design No.', 'Size']:
    if col in sale_report_df.columns:
        # 最頻値で埋める（より一般的な方法）
        mode_val = sale_report_df[col].mode()[0] # mode()はSeriesを返すため、[0]で値を取得
        sale_report_df[col].fillna(mode_val, inplace=True)
        print(f"'{col}' 列の欠損値を最頻値 '{mode_val}' で埋めました。")

        # または 'Unknown' などのプレースホルダーで埋める場合
        # sale_report_df[col].fillna('Unknown', inplace=True)
        # print(f"'{col}' 列の欠損値を 'Unknown' で埋めました。")

print("\n--- Sale Report DataFrame 欠損値処理後 ---")
display(sale_report_df.isnull().sum()[sale_report_df.isnull().sum() > 0].to_frame(name='Missing Values After'))
# 処理後のDataFrameを辞書に更新
dfs['Sale Report'] = sale_report_df
````

### 欠損値の処理（例：`Amazon Sale Report`）

`Amazon Sale Report` DataFrameには多くの欠損値があり、中には欠損率が非常に高い列もあります。このような場合、列を削除することも有効な戦略です。また、数値型とカテゴリ型で異なる処理を適用します。

-   **欠損率が高い列（`fulfilled-by`, `Unnamed: 22`, `promotion-ids`）**: 削除を検討します。
-   **カテゴリ型（`currency`）**: 最頻値または 'Unknown' で欠損値を埋めます。
-   **数値型（`Amount`）**: 中央値で欠損値を埋めます。
-   **その他（`Courier Status`, `ship-postal-code` など）**: 最頻値や 'Unknown' で埋めるか、分析の目的に応じて処理を検討します。

````python
# Amazon Sale Report DataFrameの欠損値処理
amazon_sale_df = dfs['Amazon Sale Report'].copy()

print("--- Amazon Sale Report DataFrame 欠損値処理前 ---")
display(amazon_sale_df.isnull().sum()[amazon_sale_df.isnull().sum() > 0].to_frame(name='Missing Values Before'))

# 欠損率が高い列を削除する例
# 欠損値が50%以上の列を特定して削除
threshold = 0.5 * len(amazon_sale_df)
drop_columns = amazon_sale_df.columns[amazon_sale_df.isnull().sum() > threshold].tolist()
if drop_columns:
    amazon_sale_df.drop(columns=drop_columns, inplace=True)
    print(f"欠損値が50%以上の列を削除しました: {drop_columns}")

# 数値型の欠損値を中央値で埋める (例: 'Amount')
if 'Amount' in amazon_sale_df.columns and pd.api.types.is_numeric_dtype(amazon_sale_df['Amount']):
    median_amount = amazon_sale_df['Amount'].median()
    amazon_sale_df['Amount'].fillna(median_amount, inplace=True)
    print(f"'Amount' 列の欠損値を中央値 {median_amount} で埋めました。")

# カテゴリ型の欠損値を最頻値または 'Unknown' で埋める (例: 'currency', 'Courier Status')
for col in ['currency', 'Courier Status', 'ship-state', 'ship-city', 'ship-country']:
    if col in amazon_sale_df.columns:
        mode_val = amazon_sale_df[col].mode()[0]
        amazon_sale_df[col].fillna(mode_val, inplace=True)
        print(f"'{col}' 列の欠損値を最頻値 '{mode_val}' で埋めました。")

# 'ship-postal-code' のような列は、特定のルールに基づいて埋めるか、そのまま残すかを検討
# 例: 'ship-postal-code' を0で埋める場合 (ただし、適切なロジックが必要です)
if 'ship-postal-code' in amazon_sale_df.columns and pd.api.types.is_numeric_dtype(amazon_sale_df['ship-postal-code']):
    amazon_sale_df['ship-postal-code'].fillna(0, inplace=True)
    print("'ship-postal-code' 列の欠損値を0で埋めました。")

print("\n--- Amazon Sale Report DataFrame 欠損値処理後 ---")
display(amazon_sale_df.isnull().sum()[amazon_sale_df.isnull().sum() > 0].to_frame(name='Missing Values After'))
# 処理後のDataFrameを辞書に更新
dfs['Amazon Sale Report'] = amazon_sale_df
````

### 残っている欠損値の処理（`Amazon Sale Report`）

### Amazon Sale Report の売上推移を可視化

`Amazon Sale Report` DataFrameを使用して、日付ごとの総売上を計算し、その推移を折れ線グラフで表示します。これにより、販売の傾向やパターンを把握できます。

````python
import matplotlib.pyplot as plt
import seaborn as sns

amazon_sales_df_viz = dfs['Amazon Sale Report'].copy()

# 'Date' 列を datetime 型に変換
# errors='coerce' を使用して、変換できない値をNaT（Not a Time）として処理
amazon_sales_df_viz['Date'] = pd.to_datetime(amazon_sales_df_viz['Date'], format='%m-%d-%y', errors='coerce')

# 'Amount' 列を数値型に変換
# errors='coerce' を使用して、変換できない値をNaNとして処理
amazon_sales_df_viz['Amount'] = pd.to_numeric(amazon_sales_df_viz['Amount'], errors='coerce')

# 変換後に欠損値となった行を削除
amazon_sales_df_viz.dropna(subset=['Date', 'Amount'], inplace=True)

# 日付ごとに総売上を集計
daily_sales = amazon_sales_df_viz.groupby('Date')['Amount'].sum().reset_index()

# 売上の推移を可視化
plt.figure(figsize=(14, 7))
sns.lineplot(x='Date', y='Amount', data=daily_sales)
plt.title('Amazon Daily Sales Trend')
plt.xlabel('Date')
plt.ylabel('Total Sales Amount')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

### International sale Report のデータ型変換と欠損値処理

`International sale Report` DataFrameの特定の列についてデータ型を変換し、それに伴って発生する可能性のある欠損値を処理します。具体的には、`PCS`, `RATE`, `GROSS AMT`を数値型に、`DATE`を日付型に変換します。変換後に生じた欠損値は、数値型列では中央値で、日付型列では最頻値で補完します。

**Reasoning**:
A markdown cell has already been generated to explain the upcoming data type conversions and missing value handling for the 'International sale Report' DataFrame. Now, I will add a code cell to perform these operations as per the instructions.



````python
international_sale_df = dfs['International sale Report'].copy()

print("--- International sale Report DataFrame データ型変換前 ---")
display(international_sale_df.isnull().sum()[international_sale_df.isnull().sum() > 0].to_frame(name='Missing Values Before'))

# PCS, RATE, GROSS AMT列を数値型に変換
for col in ['PCS', 'RATE', 'GROSS AMT']:
    if col in international_sale_df.columns:
        international_sale_df[col] = pd.to_numeric(international_sale_df[col], errors='coerce')
        print(f"'{col}' 列を数値型に変換しました。")

# DATE列を日付型に変換
if 'DATE' in international_sale_df.columns:
    # Explicitly specifying the date format to avoid UserWarning
    international_sale_df['DATE'] = pd.to_datetime(international_sale_df['DATE'], format='%Y-%m-%d', errors='coerce')
    print("'DATE' 列を日付型に変換しました。")

print("\n--- International sale Report DataFrame 変換後の欠損値確認 ---")
missing_values_after_conversion = international_sale_df.isnull().sum()
display(missing_values_after_conversion[missing_values_after_conversion > 0].to_frame(name='Missing Values After Conversion'))

# 数値型列の欠損値を中央値で埋める
for col in ['PCS', 'RATE', 'GROSS AMT']:
    if col in international_sale_df.columns and international_sale_df[col].isnull().any():
        median_val = international_sale_df[col].median()
        international_sale_df[col].fillna(median_val, inplace=True)
        print(f"'{col}' 列の欠損値を中央値 {median_val} で埋めました。")

# 日付型列の欠損値を最頻値で埋める
if 'DATE' in international_sale_df.columns and international_sale_df['DATE'].isnull().any():
    mode_date = international_sale_df['DATE'].mode()[0]
    international_sale_df['DATE'].fillna(mode_date, inplace=True)
    print(f"'DATE' 列の欠損値を最頻値 {mode_date.strftime('%Y-%m-%d')} で埋めました。")

# その他のカテゴリ型列の欠損値を最頻値または 'Unknown' で埋める
for col in ['Months', 'CUSTOMER', 'Style', 'SKU', 'Size']:
    if col in international_sale_df.columns and international_sale_df[col].isnull().any():
        mode_val = international_sale_df[col].mode()[0] # mode()はSeriesを返すため、[0]で値を取得
        international_sale_df[col].fillna(mode_val, inplace=True)
        print(f"'{col}' 列の欠損値を最頻値 '{mode_val}' で埋めました。")


print("\n--- International sale Report DataFrame 欠損値処理後 ---")
missing_values_final = international_sale_df.isnull().sum()
if missing_values_final.sum() == 0:
    print("このDataFrameには欠損値がありません。")
else:
    display(missing_values_final[missing_values_final > 0].to_frame(name='Missing Values After Final Processing'))

# 処理後のDataFrameを辞書に更新
dfs['International sale Report'] = international_sale_df
````

**Reasoning**:
The next step is to integrate the sales data from 'Amazon Sale Report' and 'International sale Report' into a single DataFrame. This involves adding a 'Sales Channel' column to 'International sale Report', standardizing column names like 'PCS' to 'Qty' and 'GROSS AMT' to 'Amount' in 'International sale Report' to match 'Amazon Sale Report', and then concatenating the two DataFrames.



````python
import numpy as np

# Amazon Sale Report と International sale Report を結合し、単一の売上DataFrameを作成

# Amazon Sale Report に 'Sales Channel' 列を追加
if 'Amazon Sale Report' in dfs:
    dfs['Amazon Sale Report']['Sales Channel'] = 'Amazon'
    print("Amazon Sale Report に 'Sales Channel' 列を追加しました。")

# International sale Report に 'Sales Channel' 列を追加
if 'International sale Report' in dfs:
    dfs['International sale Report']['Sales Channel'] = 'International'
    print("International sale Report に 'Sales Channel' 列を追加しました。")

# International sale Report の列名をAmazon Sale Reportと合わせる
if 'International sale Report' in dfs:
    dfs['International sale Report'] = dfs['International sale Report'].rename(columns={
        'PCS': 'Qty',
        'GROSS AMT': 'Amount'
    })
    print("International sale Report の 'PCS' を 'Qty' に、'GROSS AMT' を 'Amount' にリネームしました。")

# 結合する列を特定 (両方に存在する意味のある列)
# 'Category' は International sale Report に存在しないため、ここでは含めない
common_sales_cols = [
    'Date',
    'SKU',
    'Style',
    'Qty',
    'Amount',
    'Sales Channel',
    'Size'
]

# Amazon Sale Report と International sale Report から必要な列を抽出し、データ型を統一
amazon_sales_temp = dfs['Amazon Sale Report'][common_sales_cols].copy()
international_sales_temp = dfs['International sale Report'][common_sales_cols].copy()

# QtyとAmountのデータ型をfloatに統一（Amountはすでにfloat、QtyはIntとFloatが混じる可能性があるため）
amazon_sales_temp['Qty'] = amazon_sales_temp['Qty'].astype(float)
international_sales_temp['Qty'] = international_sales_temp['Qty'].astype(float)

# 統合された売上DataFrameを作成
combined_sales_df = pd.concat([
    amazon_sales_temp,
    international_sales_temp
], ignore_index=True)

dfs['Combined Sales'] = combined_sales_df

print("Amazon Sale Report と International sale Report を結合し、'Combined Sales' DataFrame を作成しました。")
print("\n--- Combined Sales DataFrame (head) ---")
display(dfs['Combined Sales'].head())
print("\n--- Combined Sales DataFrame (info) ---")
dfs['Combined Sales'].info()
````

**Reasoning**:
The next step is to integrate `Sale Report` and `May-2022` to create a product master DataFrame. This involves merging them on 'SKU' and 'Style' and resolving any duplicate columns by prioritizing data from `May-2022`.



````python
import numpy as np

# 3. Sale Report と May-2022 を統合し、製品マスターDataFrameを作成

# 結合キーとなるSKUとStyleが存在することを確認
if 'Sale Report' in dfs and 'May-2022' in dfs:
    # Sale ReportとMay-2022のDataFrameを取得
    sale_report_df_temp = dfs['Sale Report'].copy()
    may_2022_df_temp = dfs['May-2022'].copy()

    # 共通する列を特定し、May-2022のデータを優先するためにリネーム
    # indexは結合キーではないため無視する
    common_cols_to_rename = [
        col for col in sale_report_df_temp.columns
        if col in may_2022_df_temp.columns and col not in ['SKU', 'Style', 'index']
    ]

    # Sale Reportの重複列に接尾辞'_sales'を追加
    sale_report_df_temp = sale_report_df_temp.rename(columns={col: col + '_sales' for col in common_cols_to_rename})

    # 製品マスターDataFrameの作成
    # SKUとStyleをキーとして外部結合 (May-2022の全ての製品情報を保持しつつ、Sale Reportの情報を付与)
    # カテゴリ情報も後で活用できるように残す
    product_master_df = pd.merge(
        may_2022_df_temp,
        sale_report_df_temp,
        on=['SKU', 'Style'],
        how='left'
    )

    dfs['Product Master'] = product_master_df
    print("Sale Report と May-2022 を結合し、'Product Master' DataFrame を作成しました。")

    print("\n--- Product Master DataFrame (head) ---")
    display(dfs['Product Master'].head())
    print("\n--- Product Master DataFrame (info) ---")
    dfs['Product Master'].info()
else:
    print("'Sale Report' または 'May-2022' DataFrame がdfs辞書に見つかりませんでした。")
````

**Reasoning**:
The next step is to integrate the `Combined Sales` DataFrame with the `Product Master` DataFrame using 'SKU' and 'Style' as common keys. This will enrich the sales transactions with product-specific attributes from the master data. Duplicate columns will be handled by dropping the redundant 'Size' from the product master after the merge.



````python
import numpy as np

# 4. 統合売上DataFrameと製品マスターDataFrameを結合

if 'Combined Sales' in dfs and 'Product Master' in dfs:
    # 結合するDataFrameを取得
    combined_sales_df_temp = dfs['Combined Sales'].copy()
    product_master_df_temp = dfs['Product Master'].copy()

    # 結合キー以外の重複列を特定し、product_master_df_tempから削除する
    # combined_sales_df_tempの'Size'を残すため、product_master_df_tempの'Size'は削除
    columns_to_drop_from_product_master = [
        col for col in product_master_df_temp.columns
        if col in combined_sales_df_temp.columns and col not in ['SKU', 'Style']
    ]

    # 'index_x'や'index_y'など、結合後に不要になる可能性のある列も削除対象に含める
    columns_to_drop_from_product_master = list(set(columns_to_drop_from_product_master + ['index_x', 'index_y', 'Category_sales', 'Size']))

    # 実際に存在する列のみを削除対象とする
    columns_to_drop_from_product_master = [col for col in columns_to_drop_from_product_master if col in product_master_df_temp.columns]

    product_master_df_temp = product_master_df_temp.drop(columns=columns_to_drop_from_product_master)

    # 'SKU'と'Style'をキーとして売上データに製品属性を結合
    final_integrated_df = pd.merge(
        combined_sales_df_temp,
        product_master_df_temp,
        on=['SKU', 'Style'],
        how='left'
    )

    dfs['Final Integrated Data'] = final_integrated_df

    print("統合売上DataFrameと製品マスターDataFrameを結合し、'Final Integrated Data' DataFrameを作成しました。")
    print("\n--- Final Integrated Data DataFrame (head) ---")
    display(dfs['Final Integrated Data'].head())
    print("\n--- Final Integrated Data DataFrame (info) ---")
    dfs['Final Integrated Data'].info()
else:
    print("'Combined Sales' または 'Product Master' DataFrame がdfs辞書に見つかりませんでした。")
````

### プロジェクト概要

本プロジェクトは、`Sale Report.csv`, `Amazon Sale Report.csv`, `International sale Report.csv`, `May-2022.csv` の複数のデータソースを統合し、販売チャネル、製品カテゴリ、サイズ、および製品スタイルごとの売上パフォーマンスを分析することを目的としています。主要なトレンドの特定、懸念事項の指摘、および売上と収益性を向上させるための具体的な戦略の提案を行います。

### 使用データ

*   **Sale Report.csv**: 製品在庫と販売情報。
*   **Amazon Sale Report.csv**: Amazonチャネルからの販売トランザクションデータ。
*   **International sale Report.csv**: 海外販売チャネルからの販売トランザクションデータ。
*   **May-2022.csv**: 製品マスターデータ、価格情報、重量など。

### データ準備と統合

1.  **データ読み込み**: すべてのCSVファイルをPandas DataFrameとして読み込みました。
2.  **欠損値処理**: 各DataFrameの欠損値を確認し、数値型は中央値、カテゴリ型は最頻値で補完しました。特に、`Amazon Sale Report` では欠損値の多い列を削除しました。`International sale Report` と `May-2022` のデータ型を変換し、欠損値を適切に処理しました。
3.  **列名統一**: 複数のDataFrame間で共通のキーとなる列（`SKU Code` を `SKU`、`Design No.` や `Style Id` を `Style`、`DATE` を `Date` など）の名称を統一しました。
4.  **データ統合**:
    *   `Amazon Sale Report` と `International sale Report` を結合し、`Sales Channel` 列を追加して `Combined Sales` DataFrame を作成しました。
    *   `Sale Report` と `May-2022` を結合し、`Product Master` DataFrame を作成しました。この際、`May-2022` の情報を優先し、重複する列はリネームまたは削除しました。
    *   `Combined Sales` と `Product Master` を結合し、`Final Integrated Data` DataFrame を作成しました。この最終統合データからは、結合によって生じた100%欠損値の列を削除し、分析に適した状態にしました。

### 実施した分析

*   **販売チャネル別売上分析**: 各販売チャネル（Amazon, International）の総売上を棒グラフで可視化しました。
*   **サイズ別売上分析**: 各サイズカテゴリごとの総売上を棒グラフで可視化し、上位/下位のサイズを特定しました。
*   **トップスタイル売上分析**: 総売上に基づいてトップ10の`Style`コードを特定し、その売上を棒グラフで可視化しました。
*   **SKU/Style別売上パフォーマンス分析**: `SKU`と`Style`ごとの総販売数量、総売上金額、販売頻度を計算し、上位/下位の製品を特定しました。
*   **時系列売上予測**: 各販売チャネルの月次総売上金額の時系列データを作成し、SARIMAモデルを訓練して将来の売上を予測しました。`International`チャネルについては予測を実行し可視化しましたが、`Amazon`チャネルはデータ期間が短すぎるためスキップされました。

### 主要な分析結果と洞察

*   **販売チャネル**: `Amazon`が主要な売上貢献チャネルであり、`International`も一定の売上を上げています。ただし、`Amazon`チャネルのデータは時系列分析には不十分でした。
*   **製品パフォーマンス**: 特定の`SKU`と`Style`が売上の大部分を占めており、上位製品と下位製品の間に大きな乖離が見られます。サイズ別では、最も売上の高いサイズと低いサイズが特定されました。
*   **時系列予測 (`International`チャネル)**: SARIMAモデルは`International`チャネルの将来3ヶ月間の売上を予測しましたが、過去の変動パターンを反映しつつも、明確な長期的トレンドは示されませんでした。予測の信頼性は過去の急激な変動や外部要因に影響されます。
*   **データギャップ**: `Product Master` DataFrameとの統合後、多くの製品属性列（`Category`, `Weight`, `MRP`関連、`Color`など）が`Final Integrated Data`で100%欠損値となりました。これは、結合キーの不一致や元のデータソースに情報がなかったことが原因であり、結果として製品カテゴリごとの分析や利益計算が現在のデータセットでは不可能であるという課題が残っています。
